```python
dfrows=dfless.copy()
dfrows.head()
dfrows.count()
```

```python
dfrows = dfrows.drop(dfrows[dfrows['state']=='canceled']. index)
dfrows = dfrows.drop(dfrows[dfrows['state']=='undefined']. index)
dfrows = dfrows.drop(dfrows[dfrows['state']=='live']. index)
dfrows = dfrows.drop(dfrows[dfrows['state']=='suspended']. index)
dfrows.count()
```

```python
###Converting state to a numerical value
```

```python
dfrows['state'].unique()
```

```python
def state_to_num(x):
    if x=='successful':
        return 1
    elif x=='failed':
        return 0
    
dfrows['state_num'] = dfrows['state'].apply(state_to_num)
dfrows.tail()
```

```python
sns.countplot(x='state', data=dfrows)
```
